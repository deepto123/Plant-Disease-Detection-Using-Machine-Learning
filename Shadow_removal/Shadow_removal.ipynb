```python
import numpy as np
import cv2
from skimage.io import imsave

path = "/content/drive/MyDrive/Segmented_images/Canny_augmented_images/Apple_cedar_rust"
img_arr_org = []

for files in tqdm(glob.glob(os.path.join(path,"*.JPG"))):
  img = Image.open(files)
  img_arr_org.append(np.asarray(img))


for i, image in enumerate(img_arr_org):
  blank_mask = np.zeros(image.shape, dtype=np.uint8)
  original = image.copy()
  hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
  lower = np.array([10, 95, 105])#55,64
  upper = np.array([200, 255, 255])
  mask = cv2.inRange(hsv, lower, upper)

  cnts = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
  cnts = cnts[0] if len(cnts) == 2 else cnts[1]
  cnts = sorted(cnts, key=cv2.contourArea, reverse=True)
  for c in cnts:
      cv2.drawContours(blank_mask,[c], -1, (255,255,255), -1)
      break

  result = cv2.bitwise_and(original,blank_mask)
  imsave('/content/drive/MyDrive/Segmented_images/Shadow_black_on_canny_augmented/Apple_rust/'+str(i+1)+'.JPG',result)
 


```
